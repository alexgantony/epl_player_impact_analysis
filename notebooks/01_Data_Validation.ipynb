### Processing DataFrame

```python
def preprocess_dataframe(df):
    df.rename(columns={"rating": "sofascoreRating"}, inplace=True)
    # df["dfName"] = df.attrs["name"]

    df_player_normalized = pd.json_normalize(df["player"])
    df_player_normalized.rename(columns={"id": "playerId"}, inplace=True)
    df_player_normalized = df_player_normalized[["name", "playerId"]]

    df_team_normalized = pd.json_normalize(df["team"])
    df_team_normalized.rename(columns={"name": "teamName", "id": "teamId"}, inplace=True)
    df_team_normalized = df_team_normalized[["teamName", "teamId"]]

    df = pd.concat(
    [df.drop(["player", "team"], axis=1), 
     df_player_normalized, 
     df_team_normalized], 
     axis=1)
    
    return df
```

```python
attack_df = preprocess_dataframe(attack_df)
defence_df = preprocess_dataframe(defence_df)
passing_df = preprocess_dataframe(passing_df)
goalkeeper_df = preprocess_dataframe(goalkeeper_df)
summary_df = preprocess_dataframe(summary_df)
```

```python
def add_col_suffix(df, suffix):
    new_col = []

    for col in df.columns:
        if col != "playerId":
            new_col.append(f"{col}_{suffix}")
        else:
            new_col.append(col)

    return new_col
```

```python
attack_df.columns = add_col_suffix(attack_df, "attack")
defence_df.columns = add_col_suffix(defence_df, "defence")
passing_df.columns= add_col_suffix(passing_df, "passing")
goalkeeper_df.columns = add_col_suffix(goalkeeper_df, "goalkeeper")
summary_df.columns = add_col_suffix(summary_df, "summary")
```

```python
from functools import reduce

dfs = [attack_df, defence_df, passing_df, goalkeeper_df, summary_df]
result_df = reduce(lambda left, right: pd.merge(left, right, on="playerId", how="outer"), dfs)
result_df
```
